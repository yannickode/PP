```python
# Packages
### Tkinter Packages
import tkinter as tk
from tkinter import filedialog
from tkinter.filedialog import askopenfilename
import ttkbootstrap as ttk
from ttkbootstrap.constants import *
from tkinter import *

### Matplotlib Packages
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg, NavigationToolbar2Tk

### Sklearn Packages
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
scaler = MinMaxScaler()
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report 

### Scipy Packages
from scipy.special import expit

### Other Packages
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg, NavigationToolbar2Tk
from matplotlib.figure import Figure
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import string
import os
```

```python
def csvFile_import():
    global num_cols, cat_cols, axis, df
    pathFile = askopenfilename(title="Open File", filetypes=[("Files", "*.csv")])         # Find csv file
    # Import Data
    df = pd.read_csv(pathFile)  
    df = df.dropna()
    num_cols = df.select_dtypes(include=['number']).columns
    cat_cols = df.select_dtypes(exclude=['number']).columns
    print(num_cols)
    print(cat_cols)
    

    # axis = tuple(num_cols+cat_cols)
def plotting():
    fig = Figure(figsize=(5, 5), dpi=100)
    last_numeric_list = num_cols
    for i, col in enumerate(last_numeric_list):
        if i >= 1:
            xdata = df[last_numeric_list[i]]
            ydata = df[last_numeric_list[i-1]]

            xdata = np.array(xdata).reshape(-1,1)
            ydata = np.array(ydata).reshape(-1,1)

            xdata = scaler.fit_transform(xdata)
            ydata = scaler.fit_transform(ydata)

            plot1 = fig.add_subplot()
            plot1.scatter(xdata,ydata)

            canvas = FigureCanvasTkAgg(fig, master=window)
            canvas.draw()
            canvas.get_tk_widget().pack()

            toolbar = NavigationToolbar2Tk(canvas, window)
            toolbar.update()
            canvas.get_tk_widget().pack()

def analyze_plot():
    xdata = []                                                      # Normalization
    ydata = []
    last_numeric_list = num_cols
    for i, col in enumerate(last_numeric_list):
        if i >= 1:
            xdata = df[last_numeric_list[i]]
            ydata = df[last_numeric_list[i-1]]
            xdata = np.array(xdata).reshape(-1,1)
            xdata = scaler.fit_transform(xdata)

            x_train, x_test, y_train, y_test = train_test_split(xdata,ydata,test_size = 0.3,random_state=23)
            lr = LinearRegression() 
            lr.fit(x_train, y_train) 
            b = lr.intercept_ 
            m = lr.coef_
            # y_pred_train = m*x_train + b
            y_pred_train = lr.predict(x_train)

            plt.scatter(xdata, ydata)
            plt.plot(x_train, y_pred_train, color = "red")
            plt.xlabel(f'{last_numeric_list[i]}')
            plt.ylabel(f'{last_numeric_list[i-1]}')
            plt.legend(["Data","Best fit"])
            plt.show()
    print(f'R Squared is {lr.score(x_test,y_test)}')
    print(f'Root Mean Squared Error is {np.sqrt(mean_squared_error(x_test,y_test))}')
    print(f'Sum of Squared Error is {np.sum((lr.predict(x_test) - y_test) ** 2)}')
    std_xtest = (np.std(x_test) ** 2)
    std_ytest = (np.std(y_test) ** 2)
    deg_of_freedom = (std_xtest / len(x_test) + std_ytest / len(y_test)) ** 2 / ((std_xtest / len(x_test)) ** 2 / (len(x_test) - 1) + (std_ytest / len(y_test)) ** 2 / (len(y_test) - 1))
    print(f'Degrees of Freedom (t distribution) is {deg_of_freedom}')
    adj_rsquared = 1 - (1 - lr.score(x_test, y_test)) * (len(y_test) - 1) / (len(y_test) - x_test.shape[1] - 1)
    print(f'Adjusted R-Squared is {adj_rsquared}')

def correlation_plot():
    last_numeric_list = num_cols
    picked = [0]*len(last_numeric_list)
    highest_corr = {}
    for i, col in enumerate(last_numeric_list):
        picked[i] = col
        if i >= 1:
            newd = df.loc[:,[picked[i-1], picked[i]]]                   # To get correlation stats, only works for 2 DATA at time
            newd = newd.corr()                                          # Correlation
            plt.figure(figsize=(5,2.5))                                 # CORRELATION PLOT
            heatmap = sns.heatmap(newd, vmin=-1, vmax=1, annot=True)
            title = 'Correlation Heatmap'
            heatmap.set_title(title, fontdict={'fontsize':12},pad=12)
            plt.show()
            highest_corr[picked[i-1] +" and "+ picked[i]] = newd.iloc[:,0][1]
        else: continue
    print(f'Highest correlation is: {highest_corr}')

def polyno_plot():
    last_numeric_list = num_cols
    for i, col in enumerate(last_numeric_list):
        if i >= 1:
            xdata = df[last_numeric_list[i]]
            ydata = df[last_numeric_list[i-1]]

            xdata = np.array(xdata).reshape(-1,1)
            ydata = np.array(ydata).reshape(-1,1)

            xdata = scaler.fit_transform(xdata)
            ydata = scaler.fit_transform(ydata)

            x_list, y_list = [], [] # Convert to list
            for i, j in zip(xdata, ydata):
                x_list.extend(list(i))
                y_list.extend(list(j))

                degree = int(input("Enter the degree: "))
                degree_index = (degree+1)
                poly_list = [0]*degree_index
                for i in range(degree_index): # Get coefficients
                    poly_list[i] = np.polyfit(x_list, y_list, degree)[i]

                matrix_x = [[val**i for val in xdata] for i in range(degree_index)] # Generate Polynomial Matrix depending on the degree selected
                matrix = np.transpose(matrix_x)           # Transpose 

                poly_list_inv = poly_list[::-1]           # Reverse

                matrix_y = np.dot(matrix, poly_list_inv)  # Dot Product To get predicted y values
                matriy  = np.transpose(matrix_y)          # Transpose for proper (n X degree)

                fig, ax1 = plt.subplots()                 # Graph: Plot xdata vs predicted matrix y
                ax1.scatter(xdata, ydata)
                ax1.plot(xdata, matriy,'r')
                ax1.set_title("{} vs {}".format(num_cols[i], num_cols[i-1]))
                ax1.set_xlabel("{}".format(num_cols[i]))
                ax1.set_ylabel("{}".format(num_cols[i-1]))

                plt.show()
                

# Create Window 
window = ttk.Window(themename="journal")
window.title("Analysis App")
window.geometry('1000x700')

# Create Label for File
label_file = ttk.Label(master = window, text='Path File')    # text is Label Name
label_file.pack() # row=0, column=0 Will show Label text Name on Window
# Create String Variable
var1 = ttk.StringVar()
# Create Entry 
entry1 = ttk.Entry(master = window, textvariable=var1)    # Will Get value for use
entry1.pack() # row=0, column=1 Will show Entry on Window at row 0 and column 3
# Create Button 
button1 = ttk.Button(master = window, text='Browse CSV Data', command=csvFile_import) # Will execute csvFile_import FUNCTION
button1.pack(pady=5) # row=0, column=3 Will show Button on Window at row 0 and column 3

# Create Frame
frame_contains = ttk.Frame(window)
frame_contains.pack()

# First LabelFrame
info_frame = ttk.LabelFrame(frame_contains,
                                 text = "Data Info")
info_frame.grid(row=0, column=0)
# Analyze
button_ana = ttk.Button(info_frame,
                             text = "Linear Regression", command=analyze_plot)
button_ana.grid(row=0, column=0)
# Polynomial
button_poly = ttk.Button(info_frame,
                             text = "Polynomial", command=polyno_plot)
button_poly.grid(row=0, column=1)
# Correlation
button_corr = ttk.Button(info_frame,
                             text = "Correlation", command=correlation_plot)
button_corr.grid(row=0, column=2)

# Axis
axis = ('x-axis','y-axis')
data_cols_name1 = ttk.StringVar(value=list(axis))
data_cols_name2 = ttk.StringVar()
# 1___________________________
label_spin1 = ttk.Label(info_frame, text="X-axis")
combi1 = ttk.Combobox(info_frame)
combi1.configure(values = axis)
label_spin1.grid(row=1, column=0)
combi1.grid(row=2, column=0)
# 2___________________________
label_spin2 = ttk.Label(info_frame, text="Y-axis")
combi2 = ttk.Combobox(info_frame)
combi2.configure(values = axis)
label_spin2.grid(row=1, column=1)
combi2.grid(row=2, column=1)
# Create a Spinbox
label_spin3 = ttk.Label(info_frame, text="Degree")
spinibox = ttk.Spinbox(info_frame, from_ = 0, to = 100)
label_spin3.grid(row=1, column=2)
spinibox.grid(row=2, column=2)


for widgety in info_frame.winfo_children():
    widgety.grid_configure(padx=10, pady=5)

# Second LabelFrame
info_f1 = ttk.LabelFrame(frame_contains)
info_f1.grid(row=3, column=0, sticky="news", padx=20, pady=20)

info_f1_label = ttk.Label(info_f1, text = "Graph Info")
button_plot = Button(window, command=plotting, height=1, width=9, text="Plot Data")
button_plot.pack()



window.mainloop()                               # Continuous Loop
```
